```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
import random
from sklearn import metrics
from sklearn.linear_model import LinearRegression
dataset=datasets.load_diabetes()

print(type(dataset.data))
print(type(dataset.target))
print((dataset.feature_names))
data=dataset.data
target=dataset.target

holdout=int(len(data)*0.7)
#print(holdout)
x_training=data[:holdout]
y_training=target[:holdout]

x_testing=data[holdout:]
y_test=target[holdout:]

print(len(target))
print(len(data))
print(len(x_training))
print(len(y_training))

print(len(x_testing))
print(len(y_test))

model=SVR(kernel='linear')
model.fit(x_training, y_training)
predict=model.predict(x_testing)

print("Mean absolute error =", round(metrics.mean_absolute_error(y_test, predict), 2)) 
print("Mean squared error =", round(metrics.mean_squared_error(y_test, predict), 2)) 
print("Median absolute error =", round(metrics.median_absolute_error(y_test, predict), 2)) 
print("Explain variance score =", round(metrics.explained_variance_score(y_test, predict), 2))
print("R2 score =", round(metrics.r2_score(y_test, predict), 2))


model=LinearRegression()


model.fit(x_training, y_training)
predict=model.predict(x_testing)

print("Mean absolute error =", round(metrics.mean_absolute_error(y_test, predict), 2)) 
print("Mean squared error =", round(metrics.mean_squared_error(y_test, predict), 2)) 
print("Median absolute error =", round(metrics.median_absolute_error(y_test, predict), 2)) 
print("Explain variance score =", round(metrics.explained_variance_score(y_test, predict), 2))
print("R2 score =", round(metrics.r2_score(y_test, predict), 2))


bp=[value[3] for value in data]
print(bp)
```
